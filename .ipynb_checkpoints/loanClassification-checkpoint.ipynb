<h1>Now handeling outilers for numerical data</h1>

```python
numerical_data = ['Income','CCAvg']
```

```python
for column in numerical_data:
    sns.boxplot(df[column])
    plt.title(f'Boxplot of {column}')
    plt.show()
```

```python
#removing outilers
def outlier_cap(column):
    percentile25 = df[column].quantile(0.25)
    percentile75 = df[column].quantile(0.75)
    iqr = percentile75 - percentile25
    upper_limit = percentile75 + 1.5 * iqr
    lower_limit = percentile25 - 1.5 * iqr
    df[column] = np.where(
        df[column] > upper_limit,
        upper_limit,
        np.where(
            df[column] < lower_limit,
            lower_limit,
            df[column]
        )
    )
```

```python
outlier_cap('Income')
outlier_cap('CCAvg')
```

```python
for column in numerical_data:
    sns.boxplot(df[column])
    plt.title(f'Boxplot of {column}')
    plt.show()
```

```python
df= pd.get_dummies(df, columns=['Gender'],dtype=int)
```

```python
df.head()
```

```python
df[df['Home Ownership']=='Home Mortage']
```

```python
size_mapping = {'Rent': 0, 'Home Mortage': 1, 'Home Owner': 2}
```

```python
df['Home Ownership'] = df['Home Ownership'].map(size_mapping)
```

```python
df.head()
```

```python
df.shape
```

```python
df = df.drop(columns='ID',axis=1)
```

```python
not_approved = df[df['Personal Loan']==0]
approved = df[df['Personal Loan'] == 1]
```

```python
not_approved.shape
```

```python
approved.shape
```

```python
#since there is inbalance in dataset we have to make sure that both 0 and 1 have equal data so model will not lean to one side only
not_approved_sample = not_approved.sample(n=480)
```

```python
not_approved_sample.shape
```

```python
new_df = pd.concat([not_approved_sample,approved],axis=0)
```

```python
new_df.head()
```

```python
new_df.shape
```

```python
new_df['Personal Loan'].value_counts()
```

```python
x = new_df.drop(columns=["Personal Loan"],axis=1)
y = new_df['Personal Loan']
```

```python
x.shape
```

```python
y.shape
```

```python
#spliting train and test data
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,stratify=y)
```

```python
scaler = StandardScaler()
```

```python
#scaling the data so the whole dataset lies on in same range
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
#using logistic Regression model for thie classification problem
model = LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
x_train_predict = model.predict(x_train)
acc = metrics.accuracy_score(x_train_predict,y_train)
print(acc)
```

```python
x_test_predict = model.predict(x_test)
acc = metrics.accuracy_score(x_test_predict,y_test)
print(acc)
```

```python
print(classification_report(y_test,x_test_predict))
```

```python
print(confusion_matrix(y_test,x_test_predict))
```

```python
with open('loan_model.pkl', 'wb') as f:
    pickle.dump(model, f)
```
